# Used to test the fish subcomponent #

```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt

from aquaponics import Aquaponics
```

```python
a = Aquaponics('fish')
m = a.get_model()

tf = 15
steps = tf * 10 + 1
m.time = np.linspace(0,tf,steps)
m.options.IMODE = 4

m.solve(disp=False)
```

```python
%matplotlib inline
plt.figure(figsize=(12,6))
ax = plt.subplot(211)
plt.plot(m.time, a.FB, label='Biomass')
plt.ylabel('Fish Biomass (kcal/tank)')
ax.legend(loc=2)
ax2 = ax.twinx()
plt.plot(m.time, a.FP, label='Population', color='r')
plt.grid()
plt.legend()
plt.ylabel('Fish Population')
ax2.legend(loc=1)

ax = plt.subplot(212, sharex=ax)
plt.plot(m.time, a.INC)
plt.grid()
plt.ylabel('Nitrogen Concentration (mg / l)')

plt.xlim(0, tf)
plt.xlabel('Time (days)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

from aquaponics import Aquaponics
imode = 7
```

## 1 Plant - No Nitrogen ##

```python
a = Aquaponics('hydroplant', 'hydronitrogen', kswitch=100)
m = a.get_model()

tf = 45
steps = tf * 10 + 1
m.time = np.linspace(0,tf,steps)

a.solve(glamdring=True, imode=imode, disp=False)
```

```python
%matplotlib inline
plt.figure(figsize=(12,9))
ax = plt.subplot(311)
plt.plot(m.time, a.w, label='Dry Weight')
plt.grid()
plt.legend()

plt.subplot(312, sharex=ax)
plt.plot(m.time, a.dNup, label='dNup')
plt.grid()
plt.legend()

plt.subplot(313, sharex=ax)
plt.plot(m.time, a.cN, label='cN')
plt.grid()
plt.legend()


plt.xlim(0, tf)
plt.xlabel('Time (days)')
```

## 1 Plant - Nitrogen with Fertilizer ##

```python
a = Aquaponics('hydroplant', 'hydronitrogen', N0=0, kswitch=100)
m = a.get_model()

tf = 45
steps = tf * 10 + 1
m.time = np.linspace(0,tf,steps)
dNadd = np.ones(steps) * 0.05
a.dNadd.value = dNadd

a.solve(glamdring=True, imode=imode, disp=False)
```

```python
%matplotlib inline
plt.figure(figsize=(12,9))
ax = plt.subplot(311)
plt.plot(m.time, a.w, label='Dry Weight')
plt.grid()
plt.legend()

plt.subplot(312, sharex=ax)
plt.plot(m.time, a.dNup, label='dNup')
plt.grid()
plt.legend()

plt.subplot(313, sharex=ax)
plt.plot(m.time, a.cN, label='cN')
plt.grid()
plt.legend()


plt.xlim(0, tf)
plt.xlabel('Time (days)')
```

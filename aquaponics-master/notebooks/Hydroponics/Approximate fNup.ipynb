```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def _act(nup_w):
    if nup_w < 0.02:
        return 0
    elif nup_w > 0.05:
        return 1
    return 100 / 3 * (nup_w - 0.02)

def _approx(nup_w):
    return 1 / (1 + np.exp(-150 * (nup_w - 0.035)))
```

```python
step = 0.001
nup_ws = np.arange(0, 0.1 + step, step)
act = [_act(nup_w) for nup_w in nup_ws]
approx = [_approx(nup_w) for nup_w in nup_ws]

plt.plot(nup_ws, act, label='Modeled f(Nup)')
plt.plot(nup_ws, approx, label='Approximated f(Nup)')
plt.xlim(0, 0.1)
plt.grid()
plt.legend()
plt.xlabel('Nup / w')
plt.ylabel('f(Nup)')
```

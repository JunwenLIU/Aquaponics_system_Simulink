```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def _A(w):
    if w < 2.4:
        return 0.02 * w
    else: 
        return 0.02 * 2.4
    
def _switch(left, right, on, loc, k):
    sig = 1 / (1 + np.exp(-k * (on - loc)))
    return (1 - sig) * left + sig * right
```

```python
ws = np.arange(0, 30, 0.1)
a1 = [_A(w) for w in ws]
a2 = [_switch(0, 0.02 * 2.4, w, 1.2, 2) for w in ws]

plt.figure()
plt.plot(ws, a1, label='Modeled A')
plt.plot(ws, a2, label='Approximated A')
plt.xlim(0, 30)
plt.legend()
plt.grid()
plt.xlabel('w')
plt.ylabel('A')
```

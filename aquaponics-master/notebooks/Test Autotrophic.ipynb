# Test Fish Autotrophic Nutrient Sector #

Primarily used for debugging and making sure module runs fine on its own.

```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt

from aquaponics import Aquaponics
```

```python
a = Aquaponics('autotrophic')
m = a.get_model()

tf = 30
steps = tf * 10 + 1
m.time = np.linspace(0, tf, steps)
m.options.IMODE = 4
m.options.SOLVER = 3

a.solve(disp=False, imode=7)
```

```python
%matplotlib inline
plt.figure(figsize=(12,6))
plt.title('Autotrophic Food Quantity')
ax = plt.subplot(211)
plt.plot(m.time, a.AFe)
plt.grid()
plt.ylabel('Energy (kcal / pond)')

ax = plt.subplot(212, sharex=ax)
plt.plot(m.time, a.AFp)
plt.grid()
plt.ylabel('Protein (g protein / pond))')

plt.xlim(0, tf)
plt.xlabel('Time (days)')
```

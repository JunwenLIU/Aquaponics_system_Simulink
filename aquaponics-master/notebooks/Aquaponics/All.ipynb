```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

from aquaponics import Aquaponics
imode = 4
```

```python
a = Aquaponics(
    'all',
    ppb0=10, FP_0=10, beds=[(0, 30), (15, 45)],
    Cm_0=.6, Cb_0=.1, T0=25, kswitch=10
)
m = a.get_model()

tf = 15
steps = tf * 2 + 1
m.time = np.linspace(0,tf,steps)

start = time()
a.solve(glamdring=True, imode=imode, disp=True)
print('Solved in {:.2f} Seconds'.format(time() - start))
```

```python
%matplotlib inline
plt.figure(figsize=(12,18))
ax = plt.subplot(611)
plt.plot(m.time, a.w, label='Dry Weight')
plt.grid()
plt.legend()

plt.subplot(612, sharex=ax)
plt.plot(m.time, a.dNup, label='dNup')
plt.grid()
plt.legend()

plt.subplot(613, sharex=ax)
plt.plot(m.time, a.NH3, label='NH3')
plt.plot(m.time, a.NO2, label='N02')
plt.plot(m.time, a.NO3, label='NO3')
plt.grid()
plt.legend()
plt.ylabel('Concentration (mg/l)')

plt.subplot(614, sharex=ax)
plt.plot(m.time, a.Cm, label='Cm')
plt.plot(m.time, a.Cb, label='Cb')
plt.grid()
plt.legend()
plt.ylabel('Concentration (mg/l)')

ax1 = plt.subplot(615, sharex=ax)
plt.plot(m.time, a.FB, label='Biomass')
plt.ylabel('Fish Biomass (kcal/tank)')
ax.legend(loc=2)
ax2 = ax1.twinx()
plt.plot(m.time, a.FP, label='Population', color='r')
plt.grid()
plt.legend()
plt.ylabel('Fish Population')
ax2.legend(loc=1)

ax = plt.subplot(616, sharex=ax)
plt.plot(m.time, a.INC)
plt.grid()
plt.ylabel('Nitrogen Concentration (mg / l)')

plt.xlim(0, tf)
plt.xlabel('Time (days)')
```

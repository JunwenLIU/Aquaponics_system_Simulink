```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Outer: Fish Stocking
# Inner: Plants
weight = {
    10: {
        0: 0,
        1: 173.2,
        5: 866.0,
        10: 1.732e3,
        20: 3.371e3,
        30: 4.314e3,
        40: 4.975e3,
        50: 5.445e3,
        60: 5.849e3,
        70: 6.046e3,
        80: 6.182e3,
        90: 6.290e3,
        100: 6.385e3,
        150: 6.674e3,
        200: 6.829e3,
    }
}
mortality = {
    10: {
        0: 6,
        1: 6,
        5: 6,
        10: 6,
        20: 6,
        30: 6,
        40: 6,
        50: 6,
        60: 6,
        70: 6,
        80: 6,
        90: 6,
        100: 6,
        150: 6,
        200: 6
    }
}
biomass = {
    10: {
        0: 618.8,
        1: 2.290e3,
        5: 3.124e3,
        10: 3.262e3,
        20: 3.468e3,
        30: 3.541e3,
        40: 3.586e3,
        50: 3.609e3,
        60: 3.626e3,
        70: 3.595e3,
        80: 3.647e3,
        90: 3.653e3,
        100: 3.663e3,
        150: 3.682e3,
        200: 3.694e3
    }
}
dfweight = pd.DataFrame(weight)
dfmortality = pd.DataFrame(mortality)
dfbiomass = pd.DataFrame(biomass)
```

```python
avbiomass = dfbiomass / (10 - dfmortality)

plt.figure(figsize=(12,9))
ax1 = plt.subplot(321)
plt.plot(dfweight.axes[0], dfweight[10])
plt.ylabel('Plant Total Weight (g)')
plt.grid()

ax2 = plt.subplot(322)
ser = (dfweight[10] / dfweight.axes[0])
plt.plot(dfweight.axes[0], ser)
plt.ylabel('Average Plant Weight (g / head)')
plt.grid()

plt.subplot(323, sharex=ax1)
plt.plot(dfbiomass.axes[0], dfbiomass)
plt.ylabel('Total Fish Biomass (kcal)')
plt.grid()
plt.xlim(0, 200)
plt.xlabel('Plants Per Bed')

plt.subplot(324, sharex=ax1)
plt.plot(avbiomass.axes[0], avbiomass)
plt.ylabel('Average Fish Biomass (kcal / fish)')
plt.grid()
plt.xlim(0, 200)
plt.xlabel('Plants Per Bed')

plt.subplot(325, sharex=ax2)
plt.plot(dfmortality.axes[0], dfmortality)
plt.ylabel('Fish Mortality')
plt.grid()
plt.xlim(0, 200)
plt.xlabel('Plants Per Bed')

```

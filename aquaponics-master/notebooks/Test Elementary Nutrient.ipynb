# Test Fish Elementary Nutrient Sector #

Primarily used for debugging and making sure module runs fine on its own.

```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt

from aquaponics import Aquaponics
```

```python
a = Aquaponics('nutrient')
m = a.get_model()

tf = 30
steps = tf * 10 + 1
m.time = np.linspace(0, tf, steps)
m.options.IMODE = 4
m.options.SOLVER = 3

m.solve(disp=False)
```

```python
%matplotlib inline
plt.figure(figsize=(12,6))
plt.title('Elementary Nutrients')
ax = plt.subplot(211)
plt.plot(m.time, a.TIN, label='Total dissolved N')
plt.plot(m.time, a.TNS, label='Total N')
plt.grid()
plt.ylabel('Nitrogen (gN / pond)')
plt.legend()

ax = plt.subplot(212, sharex=ax)
plt.plot(m.time, a.TIP, label='Total dissolved P')
plt.plot(m.time, a.TPS, label='Total P')
plt.grid()
plt.ylabel('Phosphorus (gP / pond)')
plt.legend()

plt.xlim(0, tf)
plt.xlabel('Time (days)')
```
